#Sentiment Analysis

```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
result=classifier("I hate him but unfortunately now I am with him")
print(result)
```

```python
#To use a particular model
```

```python
pipeline(task="sentiment-analysis",model="facebook/bart-large-mnli")("I hate him\I don't like him")
```

```python
classifier =pipeline(task="sentiment-analysis",model="facebook/bart-large-mnli")
tasklist= ["I hate him",
              "I like icecream",
              "I am happy when I am with my family.",
              "I wanna go to Paris"]
result=classifier(tasklist)
print(result)
```

```python
classifier =pipeline(task="sentiment-analysis",model="SamLowe/roberta-base-go_emotions")
tasklist= ["I hate him",
              "I like icecream",
              "I am happy when I am with my family.",
              "I wanna go to Paris"]
result=classifier(tasklist)
print(result)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification,DistilBertTokenizer,DistilBertForSequenceClassification
```

```python
#Loading the required model
```

```python
model_name="StepanVagin/nlptown-bert-base-multilingual-uncased-sentiment-fine-tuned"
my_model=AutoModelForSequenceClassification.from_pretrained(model_name)
my_tokenizer=AutoTokenizer.from_pretrained(model_name)
```

```python
classifier=pipeline(task="sentiment-analysis",model=my_model,tokenizer=my_tokenizer)
res=classifier("I hate him so much")
print(res)
```

```python
#Tokenization deep level understanding
```

```python
from transformers import AutoTokenizer
```

```python
tokenizer=AutoTokenizer.from_pretrained("bert-base-uncased")
text="I am a Gen AI Developer"
tokens=tokenizer.tokenize(text)
print("Tokens :", tokens)
```

```python
#Creating input ids
```

```python
input_ids=tokenizer.convert_tokens_to_ids(tokens)
print("Input IDs : ",input_ids)
```

```python
#Encoding the text
```

```python
encoded_text=tokenizer.encode(text)
print("Encode Text : ",encoded_text)
#101 denotes the sos(start of sentence)
#102 denotes the eos(end of sentence)
```

```python
# Decoding the input ids
```

```python
decode_text=tokenizer.decode(input_ids)
print("Decoded text :",decode_text)
```
